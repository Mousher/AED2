### Exercicio 5. (PRÁTICA):
#### Classe Pessoa: Crie uma classe Pessoa com nome e idade. Adicione método faz_aniversario() que incrementa a idade em 1. Use __str__() para retornar algo como "<nome>, <idade> anos". Crie duas pessoas e mostre a idade de cada uma antes e depois de chamar faz_aniversario().

```python
class Pessoa:
    def __init__ (self, nome, idade):
        self.nome = nome
        self.idade = idade
    def faz_aniversario(self):
        self.idade += 1
    def __str__ (self):
        return f"{self.nome}, {self.idade} anos."

pessoa1 = Pessoa("Ana", 35)
pessoa2= Pessoa("Danilo", 32)

Aniversario = "Antes do Aniversario"
Depois = "Depois do Aniversario"
print(Aniversario)
print(pessoa1)
print(pessoa2)

pessoa1.faz_aniversario()
pessoa2.faz_aniversario()

print("-" * len(Aniversario))

print("Depois do Aniversario")
print(pessoa1)
print(pessoa2)
print("-" * len(Depois))
```

### Projeto 1: Biblioteca de Livros
#### Contexto do mundo real: Imagine um sistema para gerenciar o empréstimo de livros em uma pequena biblioteca escolar. Queremos cadastrar livros e usuários, registrar quando alguém empresta ou devolve um livro.
#### Descrição do problema: Cada Livro tem título, autor, quantidade disponível. Cada Usuário tem nome, matrícula e lista de livros emprestados. O sistema deve permitir que um usuário faça empréstimos (reduzindo a quantidade disponível) e devoluções (aumentando quantidade). Devemos evitar empréstimos se o livro estiver indisponível. Também precisamos manter os dados organizados e seguros (ex.: não permitir mexer diretamente na quantidade disponível).

```python
class Livro:
    def __init__(self, titulo, autor, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.__quantidade = quantidade

    def emprestar(self):
        if self.__quantidade > 0:
            self.__quantidade-= 1
            return True
        return False

    def devolver(self):
        self.__quantidade += 1

    def get_quantidade(self):
        return self.__quantidade


class Usuario:
    def __init__ (self, nome, matricula):
        self.nome = nome
        self.matricula = matricula 
        self.livros_emprestados = []

    def emprestar_livro(self, livro):
        if livro.emprestar():
            self.livros_emprestados.append(livro.titulo)
            print(f"{self.nome} pegou '{livro.titulo}'")
        else:
            print("Livro indisponivel!!")

    def devolver_livro(self, livro):
        if livro.titulo in self.livros_emprestados:
            livro.devolver()
            self.livros_emprestados.remove(livro.titulo)
            print(f"{self.nome} devolveu '{livro.titulo}'")
        else:
            print(f"{self.nome} não está com '{livro.titulo}'")
```

```python
livro1 = Livro("1984", "George Orwell", 2)
user1 = Usuario("Ana", 123)

user1.emprestar_livro(livro1)
user1.emprestar_livro(livro1)
user1.emprestar_livro(livro1)  # é pra dar indisponível

user1.devolver_livro(livro1)   
```

### Projeto 2: Gerenciamento de Conta Bancária com Herança
#### Contexto do mundo real: Um banco que oferece dois tipos de conta: conta corrente e conta poupança. Queremos um sistema que permita depositar, sacar e aplicar juros (no caso da poupança).
#### Descrição do problema: Precisamos modelar contas bancárias em Python de forma orientada a objetos. Teremos:
#### •Classe base Conta com atributos comuns (titular, saldo) e métodos depositar(valor) e sacar(valor).
#### •Classe ContaCorrente que herda de Conta e adiciona, por exemplo, taxa de saque (ou limite especial).
#### •Classe ContaPoupanca que herda de Conta e tem um método extra render_juros() que aumenta o saldo de acordo com uma taxa fixa anual. Também podemos manter atributos de classe (como juros = 0.05 para 5% ao ano).

```python
class Conta:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, valor):
        self.saldo += valor
        print(f"Depósito de {valor} realizado. Saldo: {self.saldo}")

    def sacar(self, valor):
        if self.saldo >= valor:
            self.saldo -= valor
            print(f"Saque de {valor} realizado. Saldo: {self.saldo}")
        else:
            print("Saldo insuficiente!")


class ContaCorrente(Conta):
    def __init__(self, titular, saldo, taxa_saque):
        super().__init__(titular, saldo)
        self.taxa_saque = taxa_saque

    def sacar(self, valor):
        total = valor + self.taxa_saque
        if self.saldo >= total:
            self.saldo -= total
            print(f"Saque de {valor} + taxa {self.taxa_saque}. Saldo: {self.saldo}")
        else:
            print("Saldo insuficiente para saque com taxa!")


class ContaPoupanca(Conta):
    juros = 0.05  # 5% ao ano (atributo de classe)

    def render_juros(self):
        ganho = self.saldo * ContaPoupanca.juros
        self.saldo += ganho
        print(f"Juros aplicados: {ganho}. Saldo: {self.saldo}")
```

```python
conta1 = ContaCorrente("Sabrina", 1000, 5)
conta2 = ContaPoupanca("Sabrina", 1000)

conta1.depositar(200)
conta1.sacar(100)

conta2.depositar(300)
```
